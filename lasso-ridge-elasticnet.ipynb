  <p>Regularization is a core concept in machine learning used to reduce overfitting ‚Äî when a model performs well on training data but poorly on unseen data. It works by adding a penalty to the model's loss function, discouraging complex models that rely too heavily on specific features or noise.</p>

  <p style="margin-top: 15px;">During training, the model minimizes the total loss:</p>

  <p style="margin: 10px 0; font-style: italic; color: #555;">
    Total Loss = Original Loss + Œª √ó Penalty
  </p>

  <ul style="padding-left: 20px; margin-top: 10px;">
    <li><strong>Original Loss:</strong> typically the mean squared error (MSE) for regression.</li>
    <li><strong>Penalty:</strong> increases as model weights grow large.</li>
    <li><strong>Œª (lambda):</strong> controls how strong the penalty is. A larger Œª encourages a simpler model.</li>
  </ul>

  <p style="margin-top: 15px;">This balance helps the model capture essential patterns while ignoring noise.</p>

  <!-- Subtitle: Types of Regularization -->
  <h2 style="
    background: #203a43;
    color: white;
    padding: 8px 16px;
    border-radius: 8px;
    font-size: 18px;
    font-weight: bold;

  ">
    ‚≠ê Types of Regularization
  </h2>

  <!-- L1 -->
  <h3 style="
    border-left: 4px solid #8360c3;
    padding-left: 10px;
    font-size: 18px;
    font-weight: bold;
    font-family: Arial, sans-serif;
    margin: 20px 0 12px 0;
    letter-spacing: 0.5px;
    text-transform: uppercase;
    color: #5e3a8a;
  ">
    1. L1 Regularization (Lasso)
  </h3>

  <p style="margin: 10px 0; font-style: italic; color: #555;">
    Loss = Original Loss + Œª √ó Œ£ |w<sub>i</sub>|
  </p>

  <ul style="padding-left: 20px; margin-top: 0;">
    <li>Encourages sparsity: some coefficients become exactly zero.</li>
    <li>Use case: Removes irrelevant features automatically ‚Äî great for high-dimensional data.</li>
  </ul>

  <!-- L2 -->
  <h3 style="
    border-left: 4px solid #8360c3;
    padding-left: 10px;
    font-size: 18px;
    font-weight: bold;
    font-family: Arial, sans-serif;
    margin: 20px 0 12px 0;
    letter-spacing: 0.5px;
    text-transform: uppercase;
    color: #5e3a8a;
  ">
    2. L2 Regularization (Ridge)
  </h3>

  <p style="margin: 10px 0; font-style: italic; color: #555;">
    Loss = Original Loss + Œª √ó Œ£ (w<sub>i</sub>)¬≤
  </p>

  <ul style="padding-left: 20px; margin-top: 0;">
    <li>Smoothly shrinks all coefficients toward zero without eliminating any.</li>
    <li>Use case: Handles multicollinearity well ‚Äî useful for datasets where features are correlated.</li>
  </ul>

  <!-- ElasticNet -->
  <h3 style="
    border-left: 4px solid #8360c3;
    padding-left: 10px;
    font-size: 18px;
    font-weight: bold;
    font-family: Arial, sans-serif;
    margin: 20px 0 12px 0;
    letter-spacing: 0.5px;
    text-transform: uppercase;
    color: #5e3a8a;
  ">
    3. ElasticNet
  </h3>

  <p style="margin: 10px 0; font-style: italic; color: #555;">
    Loss = Original Loss + Œª‚ÇÅ √ó Œ£ |w<sub>i</sub>| + Œª‚ÇÇ √ó Œ£ (w<sub>i</sub>)¬≤
  </p>

  <ul style="padding-left: 20px; margin-top: 0;">
    <li>Combines L1 and L2: enables both feature selection and coefficient shrinkage.</li>
    <li>Use case: Ideal when you have many features, some of which are correlated or irrelevant.</li>
  </ul>
</div>

  We begin by importing the essential Python libraries for data handling, visualization, and model training. These include:
  <ul style="padding-left: 20px; margin-top: 10px;">
    <li><code>pandas</code>, <code>numpy</code> for data manipulation</li>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")
```

<!-- üîÑ Data Loading & Preparation -->
<div style="
  background: linear-gradient(135deg, #2c5364, #203a43, #0f2027);

üì• Step 2: Load and Prepare the Dataset
</div>

<div style="
  background: white;
  border-radius: 12px;

  font-family: Arial, sans-serif;

  margin-bottom: 25px;
">

  We‚Äôll start by loading the <strong>Real Estate Price Prediction</strong> dataset. The target variable is <code>Y house price of unit area</code>, while the feature <code>No</code> is just a row index and can be safely dropped.

  <div style="margin-top: 15px;">
    After loading, we split the dataset into training and testing sets ‚Äî using 85% for training and 15% for evaluation.
  </div>

  <div style="margin-top: 15px;">
    Finally, we apply <strong>standardization</strong> using <code>StandardScaler</code>. This ensures all features are on the same scale, which is crucial for regularization to work effectively.
  </div>
</div>



```python
# Load data
df = pd.read_csv("/kaggle/input/real-estate-price-prediction/Real estate.csv")
X = df.drop(['Y house price of unit area', 'No'], axis=1)
y = df['Y house price of unit area']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Standardize
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

<!-- ü§ñ Step 3: Train and Evaluate Regularized Models -->
<div style="
  background: linear-gradient(135deg, #2c5364, #203a43, #0f2027);
  color: white;
  padding: 12px 24px;
  border-radius: 12px;
  font-size: 20px;

  margin: 20px 0 10px 0;
">
ü§ñ Step 3: Train and Evaluate Regularized Models
</div>

<div style="
  background: white;
  border-radius: 12px;
  border: 1px solid #e0e0e0;
  font-family: Arial, sans-serif;
  color: #333;
  padding: 18px;
  font-size: 15px;
  line-height: 1.6;
  margin-bottom: 25px;
">
  In this step, we build and train four regression models to predict real estate prices.
  <br><br>
  Each model represents a different approach to regularization:
  <ul style="padding-left: 20px; margin-top: 10px;">
    <li><strong>Linear Regression</strong>: No penalty. Fits data purely by minimizing error.</li>
    <li><strong>Ridge (L2)</strong>: Penalizes large weights using squared values. Smoothly shrinks all coefficients.</li>
    <li><strong>Lasso (L1)</strong>: Penalizes absolute values. Can shrink some coefficients to exactly zero, performing feature selection.</li>
    <li><strong>ElasticNet</strong>: A mix of both L1 and L2 ‚Äî balances sparsity and shrinkage.</li>
  </ul>

  <div style="margin-top: 12px;">
    We use <strong>cross-validation</strong> (CV) to automatically choose the best regularization strength (Œ±). This prevents overfitting and improves generalization.
  </div>

  <div style="margin-top: 12px;">
    For each model, we report:
    <ul style="padding-left: 20px; margin-top: 10px;">
      <li><strong>MSE (Mean Squared Error)</strong>: Measures average squared prediction error. Lower is better.</li>
      <li><strong>R¬≤ (R-squared)</strong>: Measures how well the model explains the variance. Closer to 1 is better.</li>
      <li><strong>Non-zero Coefficients</strong>: Tells us how many features the model actually uses. A smaller number means more regularization (especially with Lasso).</li>
      <li><strong>Optimal Alpha</strong>: The Œ± (penalty strength) chosen via cross-validation.</li>
      <li><strong>L1 Ratio</strong> (only for ElasticNet): Indicates the balance between L1 and L2 penalties.</li>
    </ul>
  </div>

  <div style="margin-top: 15px;">
    This comparison helps us understand the trade-offs: simpler models may generalize better but risk underfitting; complex models may fit too closely to training data.
  </div>
</div>


```python
# Initialize models with CV-based alpha tuning
models = {
    "Linear": LinearRegression(),
    "Ridge": RidgeCV(alphas=np.logspace(-3, 3, 50), cv=5),
    "Lasso": LassoCV(alphas=np.logspace(-3, 0, 50), cv=5, max_iter=10000),
    "ElasticNet": ElasticNetCV(alphas=np.logspace(-3, 0, 20), 
                   l1_ratio=[.1, .5, .7, .9], cv=5, max_iter=10000)
}

# Train and evaluate
results = []
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    results.append({
        "Model": name,
        "MSE": mean_squared_error(y_test, y_pred),
        "R¬≤": r2_score(y_test, y_pred),
        "Non-Zero Coefs": sum(abs(model.coef_) > 0.001),
        "Optimal Alpha": getattr(model, 'alpha_', 'N/A'),
        "L1 Ratio": getattr(model, 'l1_ratio_', 'N/A')
    })

# Performance summary
results_df = pd.DataFrame(results).set_index("Model")
print("=== Performance Comparison ===")
display(results_df.style.background_gradient(subset=["MSE", "R¬≤"], cmap='Blues_r'))
```

While Linear Regression had the weakest performance due to lack of regularization, Lasso and ElasticNet both reduced the number of features used by eliminating less important ones.

<div style="

">
  In regression models, <strong>coefficients represent the importance or weight of each feature</strong>. A larger absolute value means the feature has a stronger influence on the prediction.
  <br><br>
  This plot compares the learned coefficients across four models:
  <ul style="padding-left: 20px; margin-top: 10px;">
    <li><strong>Linear Regression</strong>: Uses all features with no shrinkage.</li>
    <li><strong>Ridge (L2)</strong>: Shrinks all coefficients but keeps them non-zero.</li>
    <li><strong>Lasso (L1)</strong>: Drives some coefficients to exactly zero (feature elimination).</li>
    <li><strong>ElasticNet</strong>: A hybrid that shrinks some and zeros out others.</li>
  </ul>
  <div style="margin-top: 10px;">
    To make the comparison meaningful, we sort the features by their absolute coefficient in the Linear model. This lets us see how much each regularized model "shrinks" the influence of each feature.
  </div>
  <div style="margin-top: 10px;">
    Coefficient shrinkage helps reduce overfitting, especially in models with many features or multicollinearity. Lasso and ElasticNet are particularly helpful when we want a simpler model by ignoring less relevant features.
  </div>
</div>


```python
coefs = pd.DataFrame({
    "Feature": X.columns,
    "Linear": models["Linear"].coef_,
    "Ridge": models["Ridge"].coef_,
    "Lasso": models["Lasso"].coef_,
    "ElasticNet": models["ElasticNet"].coef_
}).set_index("Feature")

coefs = coefs.sort_values("Linear", key=abs, ascending=False)  # Sort by importance

x = np.arange(len(coefs))
width = 0.2

plt.figure(figsize=(12, 5))
plt.bar(x - 1.5*width, coefs["Linear"], width, label='Linear', color='#1f77b4')
plt.bar(x - 0.5*width, coefs["Ridge"], width, label='Ridge', color='#ff7f0e')
plt.bar(x + 0.5*width, coefs["Lasso"], width, label='Lasso', color='#2ca02c')
plt.bar(x + 1.5*width, coefs["ElasticNet"], width, label='ElasticNet', color='#d62728')

plt.xticks(x, coefs.index, rotation=90, ha='right')
plt.axhline(0, color='black', linewidth=0.5)
plt.ylabel("Coefficient Value")
plt.title("üí° Coefficient Comparison: Regularization vs Linear")
plt.legend()
plt.tight_layout()
plt.show()
```

This highlights how regularization helps focus on the most relevant predictors by shrinking or discarding weak ones.

üìâ Step 5: Visualizing the Regularization Path
</div>

<div style="
  background: white;
  border-radius: 12px;
  border: 1px solid #e0e0e0;
  font-family: Arial, sans-serif;
  color: #333;
  padding: 18px;
  font-size: 15px;
  line-height: 1.6;
  margin-bottom: 25px;
">
  A <strong>regularization path</strong> shows how each feature's coefficient changes as the regularization strength (Œ±) increases.
  <br><br>
  At low Œ± values, regularization is weak and the model may use all features. As Œ± increases:
  <ul style="padding-left: 20px; margin-top: 8px;">
    <li><strong>Ridge</strong>: Gradually shrinks all coefficients toward zero.</li>
    <li><strong>Lasso</strong>: Forces some coefficients to zero completely (feature elimination).</li>
    <li><strong>ElasticNet</strong>: Blends both effects depending on the <code>l1_ratio</code>.</li>
  </ul>
  This path helps us see which features remain important and which are discarded as regularization becomes stronger.
</div>


```python
# Regularization path visualizer
def plot_regularization_path(model_class, title, l1_ratio=0.5):
    alphas = np.logspace(-3, 3, 50)
    coefs = []
    
    for alpha in alphas:
        if model_class == ElasticNet:
            model = model_class(alpha=alpha, l1_ratio=l1_ratio, max_iter=10000)
        else:
            model = model_class(alpha=alpha, max_iter=10000)
        model.fit(X_train_scaled, y_train)
        coefs.append(model.coef_)
    
    coefs = np.array(coefs)
    plt.figure(figsize=(8, 4))
    
    for i in range(coefs.shape[1]):
        plt.plot(np.log10(alphas), coefs[:, i], label=X.columns[i])
    
    plt.xlabel("log10(alpha)")
    plt.ylabel("Coefficient Value")
    plt.title(f"üìâ {title} Regularization Path")
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot regularization paths
plot_regularization_path(Ridge, "Ridge")
plot_regularization_path(Lasso, "Lasso")
plot_regularization_path(ElasticNet, "ElasticNet (l1_ratio=0.5)", l1_ratio=0.5)
```
